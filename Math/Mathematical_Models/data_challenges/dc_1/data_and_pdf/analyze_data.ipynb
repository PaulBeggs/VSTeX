```python
import numpy as np
import pandas as pd
import math
from typing import Dict, List
```

```python
data_path = "Data-Challenge-1.xlsx"

df = pd.read_excel(data_path)
df = df.iloc[:, [0,1,2]]

x_bins = np.arange(0, 22, 2)
y_bins = np.arange(0, 12, 2)

df['x_grid'] = pd.cut(df['x'], bins=range(0, 22, 2), right=False)
df['y_grid'] = pd.cut(df['y'], bins=range(0, 12, 2), right=False)

grid_stats = df.groupby(['x_grid', 'y_grid'], observed=False)['T(x,y)'].agg(['mean', 'count'])


final_average = grid_stats['mean'].mean()

print(f"Simple average: {df['T(x,y)'].mean():.2f}" )
print(f"Grid-based average: {final_average:.2f}" )

print("\nGrid cells with no data:")
print(grid_stats[grid_stats['count'] == 2])

grid_stats
```
